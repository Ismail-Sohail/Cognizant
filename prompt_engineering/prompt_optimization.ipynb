```python
# Different prompts are product description for iphone 16 Pro max
prompts = {
    'Basic Prompt': 'Write a description for a smartphone',
    'Refined Prompt': 'Write a product description for the latest iphone 16 pro max smartphone. Highlight its key features such as a camera, 5G connectivity, a battery, and an design. Make the description engaging and suitable for an e-commerce website.',
    "Role-Based Prompt": "You are a professional tech reviewer writing for a leading technology blog. Write a compelling and detailed review of the iphone 16 pro max smartphone, highlighting its pros, cons, and ideal user scenarios. Compare it with competing models and provide an expert recommendation."
}

# Generate responses for each prompt
for prompt_type, prompt_text in prompts.items():
    print(f"\n{prompt_type} Output:\n" + "-" * 50)
    response = generator(prompt_text, max_length=200, do_sample=True)[0]['generated_text']
    print(response)
```

```python
# From the above, We can clearly see that the basic prompt with no proper prompts gave a vague answer. However, The refined and role based prompts gave a relevant and accurate response
```

```python
from transformers import pipeline

# Load a lightweight text-generation model
generator = pipeline("text2text-generation", model="google/flan-t5-base")tiiuae/falcon-7b-instruct

# Define the original prompt
original_prompt = "Write an email inviting colleagues to a team-building event."

# Generate AI response
response1 = generator(original_prompt, max_length=150)[0]['generated_text']

# Display the AI response
print("\n🔹 AI Response (Original Prompt):\n")
print(response1)

# Initialize an evaluation dictionary
evaluation = {}

# Define the evaluation criteria
criteria = ["Relevance", "Completeness", "Clarity", "Accuracy", "Creativity", "Efficiency"]

# Ask the user to rate each criterion
print("\n📊 Please rate the AI response (1-5 scale):")
for criterion in criteria:
    while True:
        try:
            score = int(input(f"{criterion} (1-5): "))
            if 1 <= score <= 5:
                evaluation[criterion] = score
                break
            else:
                print("⚠️ Please enter a number between 1 and 5.")
        except ValueError:
            print("⚠️ Invalid input. Enter a number between 1 and 5.")

# Compute total score
total_score = sum(evaluation.values())

# Display evaluation results
print("\n📊 Evaluation Results:")
for k, v in evaluation.items():
    print(f"{k}: {v}/5")
print(f"\n🔹 Total Score: {total_score}/30")

# Refine the prompt based on evaluation feedback
improved_prompt = "Write a professional yet engaging email inviting colleagues to a fun and interactive team-building event. Include details like date, time, venue, activities planned, and RSVP instructions."

# Generate AI response for refined prompt
response2 = generator(improved_prompt, max_length=200)[0]['generated_text']

# Display the refined AI response
print("\n🔹 AI Response (Refined Prompt):")
print(response2)
```
