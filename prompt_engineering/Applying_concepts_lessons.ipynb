#Part 2: Debugging Prompt Failures

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2-medium"  # You can use other models, such as "gpt2-medium" or "gpt2-large"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Function to generate a response from the model
def generate_response(prompt, max_length=150):
    # Encode the input prompt
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate a response from the model
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.95, temperature=0.7)

    # Decode the output and return it
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Vague Prompt
vague_prompt = "Write a story about life."

# Refined Prompt
refined_prompt = """
Write a short story about a young woman named Sarah who is at a crossroads in her life.
She has just graduated from college and is uncertain about whether to take a job offer in her hometown or move to a new city for a new opportunity.
The story should focus on her internal struggle, her relationships with her friends and family, and how she ultimately makes her decision.
Include details about the setting, her emotions, and the turning point in the story.
"""

# Generate responses for both prompts
vague_response = generate_response(vague_prompt)
refined_response = generate_response(refined_prompt)

# Output the results
print("Vague Response:\n", vague_response)
print("\nRefined Response:\n", refined_response)
```

#Part 3: Evaluating Prompt Effectiveness


```python
from transformers import pipeline

# Load a lightweight text-generation model
generator = pipeline("text2text-generation", model="google/flan-t5-base")

# Define the original prompt
original_prompt = "Write an email inviting colleagues to a team-building event."

# Generate AI response
response1 = generator(original_prompt, max_length=150)[0]['generated_text']

# Display the AI response
print("\n🔹 AI Response (Original Prompt):\n")
print(response1)

# Initialize an evaluation dictionary
evaluation = {}

# Define the evaluation criteria
criteria = ["Relevance", "Completeness", "Clarity", "Accuracy", "Creativity", "Efficiency"]

# Ask the user to rate each criterion
print("\n📊 Please rate the AI response (1-5 scale):")
for criterion in criteria:
    while True:
        try:
            score = int(input(f"{criterion} (1-5): "))
            if 1 <= score <= 5:
                evaluation[criterion] = score
                break
            else:
                print("⚠️ Please enter a number between 1 and 5.")
        except ValueError:
            print("⚠️ Invalid input. Enter a number between 1 and 5.")

# Compute total score
total_score = sum(evaluation.values())

# Display evaluation results
print("\n📊 Evaluation Results:")
for k, v in evaluation.items():
    print(f"{k}: {v}/5")
print(f"\n🔹 Total Score: {total_score}/30")

# Refine the prompt based on evaluation feedback
improved_prompt = "Write a professional yet engaging email inviting colleagues to a fun and interactive team-building event. Include details like date, time, venue, activities planned, and RSVP instructions."

# Generate AI response for refined prompt
response2 = generator(improved_prompt, max_length=200)[0]['generated_text']

# Display the refined AI response
print("\n🔹 AI Response (Refined Prompt):")
print(response2)
```
