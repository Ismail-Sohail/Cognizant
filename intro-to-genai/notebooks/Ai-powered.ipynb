```python
from groq import Groq
import os

# load the API key
api_key = os.getenv("GROQ_API_KEY")
if not api_key:
    raise ValueError("API Key not found. Make sure it is set in the .env file.")

# Initialize the client with the loaded API key
client = Groq(api_key=api_key)

# Make the request
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Explain the importance of fast language models",
        }
    ],
    model="llama-3.3-70b-versatile",
)

# Print the result
print(chat_completion.choices[0].message.content)
```

```python
import os
from dotenv import load_dotenv
from groq import Groq

# Load environment variables
load_dotenv("../../.env")

# Initialize the Groq client with your API key
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

# Function for the interactive loop
def interactive_chat():
    print("Welcome to the AI Text Completion App!")
    print("You can ask the model any question. Type 'exit' to quit.")
    
    while True:
        # Accept user input
        user_input = input("You: ")
        print("User_input: ", user_input)
        
        # Exit condition
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        
        # Send the user input to the model for text completion
        try:
            chat_completion = client.chat.completions.create(
                messages=[
                    {
                        "role": "user",
                        "content": user_input,
                    }
                ],
                model="llama-3.3-70b-versatile",  # Specify the model
            )

            # Get the response from the model and print it
            model_response = chat_completion.choices[0].message.content
            print(f"AI: {model_response}")
        
        except Exception as e:
            print(f"Error: {e}")

# Run the interactive chat function
interactive_chat()
```
